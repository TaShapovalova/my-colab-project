```python
# Выделяем входные признаки (удаляем колонку с ID пользователя)
X_train = train_df.drop(columns=['id_user']).values
X_test = test_df.drop(columns=['id_user']).values

# Выделяем целевую переменную (ID пользователя)
y_train = train_df['id_user'].values
y_test = test_df['id_user'].values

# Проверим размерности матриц
print("Размер признаков обучающей выборки:", X_train.shape)
print("Размер признаков тестовой выборки:", X_test.shape)
```

```python
from sklearn.metrics import roc_curve

def calculate_metrics(model, X_test, y_test_true, target_user_id=1):
    # Получаем вероятности предсказания для всех классов
    y_preds = model.predict(X_test)

    # Индекс нужного пользователя в One-Hot матрице
    # (Если id_user начинается с 1, то индекс — это позиция в упорядоченном массиве уникальных id)
    unique_users = sorted(np.unique(y_test_true))
    user_index = unique_users.index(target_user_id)

    # Бинаризуем метки для конкретного пользователя: 1 — это он, 0 — все остальные
    y_true_binary = (y_test_true == target_user_id).astype(int)
    y_pred_probs = y_preds[:, user_index]

    # Считаем ROC-кривую: FPR (ошибка 1 рода), TPR (1 - ошибка 2 рода), пороги
    fpr, tpr, thresholds = roc_curve(y_true_binary, y_pred_probs)

    # Ищем оптимальный порог (критерий Юдена: Maximize TPR - FPR)
    optimal_idx = np.argmax(tpr - fpr)
    optimal_threshold = thresholds[optimal_idx]

    # Ошибки первого и второго рода на оптимальном пороге
    first_kind_error = fpr[optimal_idx]
    second_kind_error = 1 - tpr[optimal_idx]

    print(f"Для пользователя ID {target_user_id}:")
    print(f"Оптимальный порог (optimal threshold): {optimal_threshold:.3f}")
    print(f"Ошибка 1 рода (first kind error): {first_kind_error:.3f}")
    print(f"Ошибка 2 рода (second kind error): {second_kind_error:.3f}")

# Пример запуска для первой модели (замени target_user_id на нужного по заданию, если это не 1)
print("--- МЕТРИКИ ДЛЯ ЗАДАНИЯ 1 ---")
calculate_metrics(model1, X_test, y_test, target_user_id=1)

print("\n--- МЕТРИКИ ДЛЯ ЗАДАНИЯ 2 ---")
calculate_metrics(model2, X_test, y_test, target_user_id=1)
```

```python


import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GaussianNoise

# 1. Исправляем кодирование (делаем одинаковое число колонок для train и test)
all_y = pd.concat([pd.Series(y_train), pd.Series(y_test)])
all_y_dummies = pd.get_dummies(all_y).values

dummy_y_train = all_y_dummies[:len(y_train)]
dummy_y_test = all_y_dummies[len(y_train):]

# 2. Фиксируем случайные числа
tf.random.set_seed(777)
np.random.seed(777)

# 3. Узнаем точное количество классов
num_classes = dummy_y_train.shape[1]

# 4. Строим архитектуру из Задания 1
model1 = Sequential()
model1.add(GaussianNoise(0.15, input_shape=(144,)))
model1.add(Dense(30, activation='relu'))
model1.add(Dense(num_classes, activation='softmax'))

# 5. Компилируем модель
model1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print("Модель №1 успешно создана! Количество классов:", num_classes)
```

```python
history1history1 = model1.fit(X_train, dummy_y_train, epochs=20, batch_size=32, validation_data=(X_test, dummy_y_test), verbose=1)
```

```python
# Фиксируем случайные числа перед созданием новой модели
tf.random.set_seed(777)
np.random.seed(777)

# Строим измененную архитектуру для Задания 2
model2 = Sequential()
model2.add(GaussianNoise(0.15, input_shape=(144,)))
model2.add(Dense(30, activation='sigmoid')) # Поменяли активацию на sigmoid
model2.add(Dense(num_classes, activation='softmax'))

# Компиляция и обучение модели №2
model2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history2 = model2.fit(X_train, dummy_y_train, epochs=20, batch_size=32, validation_data=(X_test, dummy_y_test), verbose=1)
```

```python
# Фиксируем случайные числа перед созданием новой модели
tf.random.set_seed(777)
np.random.seed(777)

# Строим трехслойную архитектуру для Задания 3
model3 = Sequential()
model3.add(GaussianNoise(0.2, input_shape=(144,))) # Шум 0.2
model3.add(Dense(30, activation='sigmoid'))       # Второй слой
model3.add(Dense(10, activation='relu'))          # Третий слой
model3.add(Dense(num_classes, activation='softmax')) # Выходной слой

# Компиляция по ТЗ (бинарная кросс-энтропия) и обучение на 50 эпох
model3.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history3 = model3.fit(X_train, dummy_y_train, epochs=50, batch_size=32, validation_data=(X_test, dummy_y_test), verbose=1)
```
